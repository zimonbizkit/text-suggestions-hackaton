```python
import numpy as np
from collections import Counter
from itertools import combinations
import math
```

```python
def calculate_pmi(df, column):
    
    # Split the keywords into tokens and get all token pairs
    token_pairs = [list(combinations(str(keywords).split(), 2)) for keywords in df[column]]
    token_pairs = [pair for sublist in token_pairs for pair in sublist]  # Flatten the list

    # Count the frequency of each token and each token pair
    token_counter = Counter([token for keywords in df[column] for token in str(keywords).split()])
    pair_counter = Counter(token_pairs)

    # Total number of keywords
    total_keywords = len(df)

    # Calculate PMI for each pair
    pmi_values = {}
    for pair in pair_counter:
        px = token_counter[pair[0]] / total_keywords
        py = token_counter[pair[1]] / total_keywords
        pxy = pair_counter[pair] / total_keywords
        pmi = math.log(pxy / (px * py), 2)
        pmi_values[pair] = pmi

    return pmi_values

# Calculate PMI for the keywords



pmi_values = calculate_pmi(signals_df, 'target')

pmi_df = pd.DataFrame(list(pmi_values.items()), columns=['Token Pair', 'PMI'])

# Sort the DataFrame by PMI values in descending order
pmi_df = pmi_df.sort_values(by='PMI', ascending=False)

pmi_df.head(400)  # Display the top PMI values
```
